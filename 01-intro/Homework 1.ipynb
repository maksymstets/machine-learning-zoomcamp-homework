```python
import pandas as pd

# Load the dataset
car = pd.read_csv('car_fuel_efficiency.csv')

# Step 1: Find the median value of the horsepower column
median_hp = car['horsepower'].median()
print(f"Median horsepower: {median_hp}")

# Step 2: Calculate the most frequent value (mode) of the horsepower column
mode_hp = car['horsepower'].mode()[0]
print(f"Most frequent horsepower: {mode_hp}")

# Step 3: Fill missing values in horsepower with the most frequent value
car['horsepower'] = car['horsepower'].fillna(mode_hp)

# Step 4: Calculate the median value of horsepower again
new_median_hp = car['horsepower'].median()
print(f"New median horsepower after filling missing values: {new_median_hp}")
```

```python
import pandas as pd
import numpy as np
car = pd.read_csv('car_fuel_efficiency.csv')
car_origin = car[car['origin'] == "Asia"]
car_choosed = car_origin[['vehicle_weight', 'model_year']].head(7)
x =car_choosed.to_numpy()
XTX = x.T@x
XTX_inv = np.linalg.inv(XTX)
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
w = XTX_inv @ x.T@y
w_sum = w.sum()
print(f"Sum of the coefficients: {w_sum}")
```
